# Step 3: Clustering Analysis

```python
# Applying K-Means clustering with an arbitrary choice of 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_data)
```

```python
# Calculating silhouette score for K-Means clustering to assess quality
kmeans_silhouette = silhouette_score(scaled_data, kmeans_labels)
print(f'Silhouette Score for K-Means Clustering: {kmeans_silhouette}')

# Implementing DBSCAN clustering
dbscan = DBSCAN(eps=1.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled_data)
```

```python
# Evaluating DBSCAN clustering using silhouette score
dbscan_silhouette = silhouette_score(scaled_data, dbscan_labels)
print(f'Silhouette Score for DBSCAN Clustering: {dbscan_silhouette}')
```

# Step 4: Profiling of Clusters

```python
# Assigning the cluster labels back to the original dataset
data['Cluster_KMeans'] = kmeans_labels
data['Cluster_DBSCAN'] = dbscan_labels

# Analyzing the characteristics of each KMeans cluster
for cluster in range(4):
    print(f'\nKMeans Cluster {cluster} Profile:\n', data[data['Cluster_KMeans'] == cluster].describe())
```
